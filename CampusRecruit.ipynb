```python
dataset['ssc_b']=dataset['ssc_b'].map({'Others':1,'Central':2})
dataset.head()
```

```python
dataset['hsc_b'].unique()
```

```python
dataset['hsc_b']=dataset['hsc_b'].map({'Others':1,'Central':2})
dataset.head()
```

```python
dataset['hsc_s'].unique()
```

```python
dataset['hsc_s']=dataset['hsc_s'].map({'Commerce':1, 'Science':2, 'Arts':3})
dataset.head()
```

```python
dataset['degree_t'].unique()
```

```python
dataset['degree_t']=dataset['degree_t'].map({'Sci&Tech':1,'Comm&Mgmt':2, 'Others':3 })
```

```python
dataset.head()
```

```python
dataset['workex'].unique()
```

```python
dataset['workex']=dataset['workex'].map({'No':0,'Yes':1})
dataset.head()
```

```python
dataset['specialisation'].unique()
```

```python
dataset['specialisation']=dataset['specialisation'].map({'Mkt&HR':1,'Mkt&Fin':2})
dataset.head()
```

```python
dataset['gender'].unique()
```

```python
dataset['gender']=dataset['gender'].map({'M':1, 'F':2})
dataset.head()
```

```python
#Scaling

from sklearn.preprocessing import StandardScaler

X=dataset.drop(['status'],axis=1)
X.head()

```

```python
y=dataset['status']
print(y)
y=y.values
print(y)
```

```python
ss=StandardScaler()
X=ss.fit_transform(X)
X
```

```python
print(type(X))
print(type(y))
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test =  train_test_split(X ,y , test_size=0.2, random_state=1)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred=lr.predict(X_test)
df = pd.DataFrame({'Actual':y_test, 'Predicted': y_pred})
df
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(accuracy_score(y_test,y_pred))

print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB

nb=GaussianNB()
nb.fit(X_train, y_train)
```

```python
y_pred=nb.predict(X_test)
df = pd.DataFrame({'Actual':y_test, 'Predicted': y_pred})
df
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn= KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)
```

```python
y_pred=knn.predict(X_test)
df = pd.DataFrame({'Actual':y_test, 'Predicted': y_pred})
df
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
y_pred=dt.predict(X_test)
df = pd.DataFrame({'Actual':y_test, 'Predicted': y_pred})
df
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred=rf.predict(X_test)
df = pd.DataFrame({'Actual':y_test, 'Predicted': y_pred})
df
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.svm import SVC


sv=SVC()
sv.fit(X_train, y_train)
```

```python
y_pred=sv.predict(X_test)
df = pd.DataFrame({'Actual':y_test, 'Predicted': y_pred})
df
```
